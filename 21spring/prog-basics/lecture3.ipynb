# Объектно-ориентированное программирование

Самая распространенная парадигма
программирования. Считается, что программа
- это множество «объектов», которые
взаимодействуют между собой, посылая друг
другу сообщения с информацией.

Пример. В программе для деканата, объекты
- это студенты, преподаватели, зачетные
книжки студентов, экзамены.
сообщения: у экзамена можно спросить дату
у зачетной книжки можно попросить добавить
оценку по определенному экзамену.

Проектирование программы - определение объектов,
из которых состоит предметная область.

На практике объекты оказываются очень
абстрактными.
 - открытый файл для чтения
 - соединение с БД. Это соединение позволяет
 отправлять запросы к БД.
 - Событие сдачи экзамена студентом преподавателю
   кто, когда, кому, что сдавал, какие
   вопросы были заданы, какую оценку получил
 - Фабрика создания событий сдачи экзамена
 - ...

Объект = данные + поведение
Объект хранит данные и умеет выполнять
команды, которые изменяют эти данные,
имеют побочные эффекты, возвращают результат
Т.е. данные ~ переменные,
поведение ~ функции

Класс - описание объектов определенного типа
Объекты - конкретные реализации класса.

Пример: Класс Студент, Объект = студент
Иванов, студент Петров, ...
У одного класса может быть много объектов.
В коде программы мы описываем классы,
во время работы программы создаются объекты.

Пример:

```python
# Описываем класс. Студент это имя и курс
class Student:
    name = ""
    course = 1

#ivanov, petrov объекты
ivanov = Student()
petrov = Student()

# доступ к информации объекта через .
ivanov.name = "Иванов"
petrov.name = "Петров"
ivanov.course = 2
petrov.course = 3

print(ivanov.name)
```

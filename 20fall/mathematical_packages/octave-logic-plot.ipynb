# Logical values, logical operators

```octave
3 + 2
3 >= 2  % true = 1
3 <= 2  % false = 0
```

```octave
[1 2 3 4 5 6 7] >= [7 6 5 4 3 2 1]
[1 2 3 4 5 6 7] >= 5 % [5 5 5 5 5 5 5]
```

1 and 0 are not numbers, they are logical values that just look like numbers. But they usually convert to numbers automatically.

## logical operators

```octave
x = 1:8
4 <= x
x <= 7
4 <= x & x <= 7 # logical and
x <= 2 | x >= 5 # logical or
~ (x <= 5) # logical not
```

## The `find` function
`find` returns indices of non-zero elements in a matrix:

```octave
find([11 22 0 0 55 33 0 0 0 44])
```

This is usually used with logical values. non-zero means true values:

```octave
a = [11 22 0 0 55 33 0 0 0 44]
a == 55
find(a == 55)
find(a >= 40)
```

Let's find elements that are divisible by 3:

```octave
a = [10 20 30 40 50 60 70]
mod(a, 3)  # a remainder in division by 3
mod(a, 3) == 0
i = find(mod(a, 3) == 0)
a(i) # select elements, that are divisible by 3
a(find(mod(a, 3) == 0)) # in one line
```

```octave
a(find(mod(a, 3) == 0)) # in one line
a(mod(a, 3) == 0) # in fact, we don't need find
```

This is called "logical indexing". If index is a logical vector, then this vectors shows which element to take, and which not to take.

# Plotting
We will discuss only the `plot` function. Imagine, we want to draw several points on a plane. p1 = (1, 2), p2 = (3, 4), p3 = (5, 1):

```octave
x = [1 3 5]; # x cords of points
y = [2 4 1]; # y cords of points
plot(x, y)
```

```octave
x = [1 3 5 2];
y = [2 4 1 2];
plot(x, y, "ro")  # r means red, o means circles
```

```octave
plot([1 2 3], [1 3 2], "ro", [4 5 4], [5 2 3], "*b")  # two graphs
```

```octave
x = [1 1; 2 2; 3 3]  # two plots similtaneously. Each column of x, y is one
y = [3 4; 5 2; 6 5]  # plot
plot(x, y)
```

```octave
x = randn(1, 1000); # normal distribution N(0, 1), 1 line by 100 columns
y = randn(1, 1000); # the same
plot(x, y, "ob")
```

A plot of a usual function. For example, y1 = log(x) and y2 = log(x) / x:

```octave
x = linspace(1, 10, 100);  # 100 numbers from 1 to 10
y1 = log(x); # evaluate log of every value inside x
y2 = log(x) ./ x; # element by division
plot(x, y1, "r", x, y2, "b")
```

```octave
phi = linspace(0, 2 * pi, 100);
x = cos(phi); # applied to each element
y = sin(phi);
plot(x, y)
```
